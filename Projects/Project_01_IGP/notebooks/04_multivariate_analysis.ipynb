# Multivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/earthquake_analysis_ready.csv")

df["date"] = pd.to_datetime(df["date"], utc=True)
df["date"] = df["date"].dt.tz_convert("America/Lima")
```

```python
df.head(3)
```

## Comparisons

```python
sns.scatterplot(data=df, x="magnitude", y="depth")

plt.title("Magnitude vs Depth of Earthquakes")
plt.xlabel("Magnitude")
plt.ylabel("Depth (km)")

plt.savefig("../reports/figures/multivariate_analysis/magnitude_vs_depth.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
sns.lineplot(data=df, x="magnitude", y="depth")
plt.title("Magnitude vs Depth of Earthquakes")
plt.xlabel("Magnitude")
plt.ylabel("Depth (km)")
plt.savefig("../reports/figures/multivariate_analysis/magnitude_vs_depth_line.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
corr = df[["depth", "magnitude", "latitude", "longitude"]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap of Earthquake Features")
plt.savefig("../reports/figures/multivariate_analysis/correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```
